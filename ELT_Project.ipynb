```python
import pandas as pd
from sqlalchemy import create_engine
import Database_Config as cfg
```

```python
csv_file = "Resources/person.csv"
person_data_df = pd.read_csv(csv_file)
# Trnaform the data to remove NaN
person_data_df=person_data_df.fillna('')
person_data_df.head(50)
```

```python
csv_file = "Resources/store.csv"
store_data_df = pd.read_csv(csv_file)
store_data_df.head()
```

```python
json_file = "Resources/email.json"
email_data_df = pd.read_json(json_file)
email_data_df.head()
```

```python
json_file = "Resources/phone.json"
phone_data_df = pd.read_json(json_file)
phone_data_df.head()
```

```python
connection_string = (f"{cfg.psql['server']}://{cfg.psql['user']}:{cfg.psql['passwd']}@{cfg.psql['host']}:{cfg.psql['port']}/{cfg.psql['db']}")
engine = create_engine(connection_string)
connection = engine.connect()
```

```python
engine.table_names()
```

```python
person_data_df.to_sql(name='person', con=connection, if_exists='append', index=False)
```

```python
store_data_df.to_sql(name='store', con=connection, if_exists='append', index=False)
```

```python
email_data_df.to_sql(name='email', con=connection, if_exists='append', index=False)
```

```python
phone_data_df.to_sql(name='phone', con=connection, if_exists='append', index=False)
```
